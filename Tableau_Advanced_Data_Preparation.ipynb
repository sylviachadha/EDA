## 2. Pivoting Data

`Pivot` = rows into columns, essentially creating a crosstab or summary view, *long format to a wide format*

`Unpivot` =  columns are transformed into rows, *wide format to a long format*

In tableau select columns u want to pivot & choose pivot option.

## 4. Metadata

To get an overview of data source structure. The metadata grid allows you to get a glance of the fields you have imported, the source tables, and the remote field name (the name of the field as it is in the data source).
