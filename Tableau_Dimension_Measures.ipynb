# Navigating Tableau

Data section on left and Worksheet area on right. The data section is always broken down into 2 parts- Dimensions & Measures.

### Dimensions and Measures

Dimensions marked with Abc and Measures with # symbol.

Change data type of Date from text to date.

Dimensions & measures r 2 different roles that any data element can take. In measures tableau puts all numerical values/quantitative data while in dimensions we have all categorical/ qualitative data. You can move data from dimensions to measures or vice versa if required based on your understanding of data, generally tableau allocates correctly.

Dimensions are independent variables while measures are dependent variables (since generally measures will be grouped by some independent variable).


It is possible to change a measure to dimension.

**Data** can be used to connect to a new data source.

**Worksheet** where you draw your visualizations using the data.

**Dashboard** it is a combination of worksheets.

**Story** A story can be a combination of worksheets and dashboards.
