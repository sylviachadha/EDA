# Dashboard

## Making Dashboard Interactive using Actions

Allow user of dashboard to drill into certain parts of dashboard by just simply clicking on it.

e.g. if u want to just see france in map then the customer scatterplot will also change accordingly if dashboard interactive.

## Actions

There r 2 types of actions in tableau - filtering and highlighting. Choose  Dashboard and Actions option.

### Types of Actions

# 1. Making Action Filters

A shortcut to add actions is on plot itself, choose Use as Filter, so filter action is added.

Before and After making Dashboard Interactive.

When we choose Use as filter option, we see generated action (means the action is added automatically).

Click on Edit

- In Source we have map of europe and action is select so when u select any part in europe map, the fields (target plots) that will be effected are Custome scatterplot and map of europe.
- Source and target sheets both r same here (Dashboard1).

We can also change action to Hover

Now no need to even select by clicking, can just select by hovering and scatterplot will change accordingly.

If u want to delete automatically generated filter, under Actions click on Remove.

To generate a new filter using Actions under Dashboard. In Add actions choose filter. Choose Name e.g Select_Geography. 

### Multiple Selections

You can select multiple places together (using command) & corresponding scatter plot will be shown as below 3 items selected.

Another way to select multiple places is using rectangular selection.

Also another options are circle selection and free form selection

# 2. Making Action Highlighters

- In highlighting the dataset is actually not filtered. Highlighting does not take only selected rows & remove rest, it is still working with whole dataset, so need to take care of level of granularity of different worksheets in dashboard.

- In filtering the dataset is filtered for specific rows as per selection and other rows r temporarily deleted from the dataset.

Choose highlight in actions
