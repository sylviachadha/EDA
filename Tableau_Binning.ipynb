# Binning for Continuous Variables

By Binning we change a continuous/numerical variable to a categorical variable.

First u change Age which is measure to dimension. After changing to dimension u go for binning.

Before Binning - each discrete age represented individually

After Binning- age groups

To create size of bins=5 meaning each bin will have 5 ages e.g. 20-25. 20 will be inclusive in this band but 25 will not be inclusive in this band.
