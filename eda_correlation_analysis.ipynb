## ScatterPlots

### Extent of correlation

- In **`Perfect +ve Correlation`**, correlation value will be close to 1.In **`Perfect -ve Correlation`**, correlation value will be close to -1.
- In **`high +ve correlation`**, data points dont exactly lie on line, r bit scattered, value will be b/w 0.9 to 0.7. If  **`high -ve correlation`**,value will be b/w -0.9 to -0.7.
- In **`low +ve correlation`**, data points r even more scattered than the line of best fit, correlation value will be close to 0.5. In **`low -ve correlation`**, value around -0.5.
- If the data points r totally scattered there will be **`no correlation`** b/w variables, correlation value will be close to 0.


## Pairplots
### identify relationships among multiple numerical variables

- To identify relationships b/w pairs of multiple numerical variables we use Pairplot which is represented using multiple charts in a matrix format.
- By default only numerical columns r included in pairplot. 
- We can include categorical columns by changing the colors of category in categorical columns e.g. adding hue parameter based on category. e.g. smoker & non-smoker below.


## Correlation Matrix

- Instead of analysing relationship visually we can look at correlation matrix to look at relationship quantitatively.
- All diagnol values will always be 1 for any correlation matrix.
- Correlation matrix indicate only linear relationships b/w variables. If there r any non-linear relationships then correlation value is not the right metric.
