## Four Phases
1. Connecting and Shaping data 
2. Creating a Data Model and Table relationships
3. Analysing data with DAX Calculations
4. Visualizing data and reports

## <font color= blue>Phase 1 Connecting and Shaping data 

Get data from tables, files, cloud services and in this phase **Query Editor** is our cockpit. You should only extract the data you need.



- **Basic table transformations** - sort, datatype, add/remove rows, column etc
- **Text specific tools** like split, upper, lower etc
- **Number specific tools** like mean,median, mode etc
- **Date specific Tools** like date, year, month, earliest, latest
- **Pivoting** (Long to wide format/row to column) and **Unpivot** (Wide to long format/column to row)
- **Joins** on tables if required (not recommended), rather relationships preferred.

## <font color= blue>Phase 2 Creating a Data Model and Table relationships

- Create **table relationships** to blend all information into single normalized data model.
- Mainly data/transaction tables and lookup/master tables.
- Values in **Primary key** column will be unique while values in **foreign key** columns may be duplicate. 
- **Relationship Cardinality**, usually 1:many i.e. one instance of each primary key but potentially many instances of each foreign key. Others are 1:1, many:1, many:many.
- One way vs two way **filter**

### 2. Calculated Columns vs Measures

Symbols -> Sigma for raw field, Calculator for Measure and fx for Calculated column.

- Calclated columns live inside your table while Measures do not, they r mainly used for visualization purpose.
- Calclated columns understand row context, great for defining properties based on each row  as well as filtering data but generally useless for aggregation (SUM, COUNT etc)
- Measures are DAX formula's to generate new calculated values.

## <font color= blue> Phase 4 Visualizing data and reports

Visualization data with Reports and Dashboards. A report can have many pages just like tabs in excel.

- Cards, Singular card, multi-row card and KPI cards
- Line and Area charts for trends
- Guage chart, compare actual performance against a target, sales goals, production targets etc
- Maps
- Matrix Visuals
- Slicers, a fancy word for filter
- Treemap, to show composition
- Forecasting tools
