# Data Wrangling using Pandas

- When we want to build reports, dashboards or ML models we will be starting with raw data. 
- Raw data is usually stored in form of flat files or database tables.
- We read raw data using various tools to either visualize or build models.
- If the quality of data is bad, you would need data cleaning & data pre-processing.

## Need for Data Wrangling

## Definition of Data Wrangling

The process or steps that r used to convert raw data from its original structure to the structure that is expected by visualization tools or model building tools is called as data wrangling.

## Techniques of Data Wrangling

Depending on use, we might use 1 or multiple of these techniques-



### 1. Data Cleaning

- Most of real time data will have missing values however many ML algorithms do not take data with missing values.
- We need to treat missing values properly before we build any ML model.
- Text data cleaning e.g. data that comes from social media platforms or scraped from websites or data entered through web-forms will have a lot of text columns with informal language. These informal text is v.difficult for computers & algorithms to interpret & hence these text data fields require a lot of data cleaning.
- Also many broad data quality issues will be encountered while cleaning data.


### 2. Summarization 

- We might need to convert data from one granularity e.g. day wise to another granularity e.g. year wise like if we want to understand sales yearly however data is given per day.
- This process in which we change granularity of data is summarizing.


### 3. Reshaping / Pivoting

- Usually data will be stored in Long format i.e. more rows & fewer columns however this is not a human readable format. 
- To make data readable we need to reshape them into a wide format. This process is called as reshaping or pivoting.



### 4. Merge Data Source

- Sometimes data might be coming from multiple sources & each of this data would be in different granularity.
- We might have to join them together before we process them.
