```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating daily price changes
daily_changes = np.array([0.50, -0.30, 0.20, 0.10, -0.50, 0.60, -0.10, 0.40, -0.20, 0.30])

# Calculating cumulative sum of daily changes
cumulative_price_changes = daily_changes.cumsum()

# Plotting the cumulative sum
plt.plot(cumulative_price_changes)
plt.title('Cumulative Stock Price Changes Over 10 Days')
plt.xlabel('Days')
plt.ylabel('Cumulative Change in Price ($)')
plt.grid(True)
plt.show()
```

## 5. Boxplot (distribution)

```python
# List comprehension syntax-

# [expression for item in iterable]   or
# [expression for item in iterable if condition]
```

```python
# Create a list of squares for numbers from 0 to 9.
# [expression for item in iterable]

squares = [num*num for num in range(0,10)]
squares
```

```python
# Create a list of even numbers from 0 to 9
# [expression for item in iterable if condition]

even = [num for num in range(0,10) if num%2==0]
even
```

```python
# List comprehension method

data = [np.random.normal(0, std, 100) for std in range(1,3)]
data[0]  # std dev 1
```

```python
data[1]    # std dev 2
```

```python
plt.boxplot(data, vert=True);
```

Above 1st plot shows most data concentrated b/w 1 and -1 as std dev 1 for this and 2nd plot shows most data concentrated b/w 2 and -2 as std dev 2 for this plot. Hence 2nd plot is more spread out cz its std deviation is more.

```python
plt.boxplot(data, vert=True, patch_artist= True);
```

### xlim, ylim, xticks, yticks

xlim and ylim are used to set or get the limits (range of data that will be displayed along an axis) of the x-axis and y-axis. This can be helpful to Focusing on a specific range of data, also for ensuring consistency across multiple plots.

xticks and yticks are used to set or get the locations and labels of the ticks on the x-axis and y-axis, respectively.

## 8. Nested Pie Chart/ Multi-level pie charts


/*<font color=blue>Pie Charts show only proportions while Nested pie chart show bith hierarchy and proportions*/

These are used to show hierarchical relationships and proportions of a dataset. Nested pie charts are not suitable for data with a large number of categories or subcategories, as they can become visually overwhelming. Alternative - Stacked Bar Chart


Below code will generate a nested pie chart with two levels: the outer ring representing the main product categories and the inner ring representing the subcategories within each main category.








```python
import matplotlib.pyplot as plt

# Data
labels = ['Fruits', 'Vegetables', 'Dairy']
sizes = [215, 130, 245]
sub_labels = ['Apple', 'Banana', 'Tomato', 'Onion', 'Milk', 'Cheese']
sub_sizes = [70, 145, 80, 50, 150, 95]

# Colors
colors = ['#ff9999','#66b3ff','#99ff99']
sub_colors = ['#c2c2f0','#ffb3e6', '#c4e17f', '#76d7c4', '#f7c6c7', '#f7b7a3']

# Pie Chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, radius=1.3, colors=colors, autopct='%1.1f%%', pctdistance=0.85, startangle=90)
ax.pie(sub_sizes, labels=sub_labels, radius=1, colors=sub_colors, autopct='%1.1f%%', pctdistance=0.75, startangle=90)
ax.set(aspect="equal")
plt.show()
```

## 9. Donut Chart

```python
import matplotlib.pyplot as plt

# Data
group_names = ['Fruits', 'Vegetables', 'Dairy']
group_sizes = [215, 130, 245]
group_colors = ['#ff9999','#66b3ff','#99ff99']

# Donut Chart
fig, ax = plt.subplots()
ax.pie(group_sizes, labels=group_names, colors=group_colors, autopct='%1.1f%%',
       startangle=90, counterclock=False, wedgeprops=dict(width=0.4))
ax.set(aspect="equal")
plt.show()
```

## 10. Bar Chart on Polar axis

Polar Axes:
- Description: Uses a radius and angle to define a point.
- Use: Useful in situations where circular symmetry is present, such as in trigonometry and physics.
- Components: Radial distance (r), Angle (θ)

Cylindrical Axes:
- Description: A 3D extension of polar coordinates.
- Components: Radial distance (r), Angle (θ), Height (z)


*Using a polar bar chart can make cyclic patterns in the data immediately apparent, which might be less obvious in a traditional linear bar chart.*








```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
hours = np.arange(0, 24, 1)  # 24 hours
visits = np.random.randint(50, 200, size=24)  # Random visit data

# Convert hours to radians
theta = np.deg2rad(hours * 15)

# Create the polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
bars = ax.bar(theta, visits, width=np.deg2rad(15))

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
hours = np.arange(0, 24, 1)  # 24 hours
visits = np.random.randint(50, 200, size=24)  # Random visit data

# Convert hours to radians
theta = np.deg2rad(hours * 15)

# Create the polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
bars = ax.bar(theta, visits, width=np.deg2rad(15))

# Set the labels for the time (hours)
# We'll label every hour, you can adjust this to label fewer points
ax.set_xticks(np.deg2rad(hours * 15))  # Position the ticks at each hour
ax.set_xticklabels([f'{h}:00' for h in hours])  # Format labels as 'hour:00'

# Display the plot
plt.show()
```
