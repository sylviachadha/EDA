### Matplotlib Object Hierarchy

There is an Object Hierarchy within Matplotlib. 

- A Figure object is the outermost container for a Matplotlib plot. The Figure object contain multiple Axes objects. So, the Figure is the final graphic that may contain one or more Axes. **The Axes represent an individual plot**

- The Axes object contain smaller objects such as tick marks, lines, legends, title and text-boxes.

```python
# create a plot figure
plt.figure()

# create the first of two panels and set current axis
# grid 2*1=2 plots and (2,1,1) is 1st plot/panel & (2,1,2) is 2nd plot/panel
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))
plt.title('2*1 grid and 1st panel')

# create the second of two panels and set current axis
plt.subplot(2, 1, 2)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1))
plt.title('2*1 grid and 2nd panel')
plt.subplots_adjust(hspace=0.4);

```

#### plt.plot() Vs plt.scatter() functions
- For smaller datasets, it does not matter whether we use plt.plot() or plt.scatter() functions to create scatter-plots.

- But for larger datasets, plt.plot() function is more efficient than plt.scatter() & takes less time/efficient considering performance point of view
- For customizations and advanced uses plt.scatter is better considering flexibility point of view.

#### Proof that plt.plot() is efficient in performance than plt.scatter() using timeit function

```python
# timeit is used to measure the execution time for the small python code snippets. 
# This module runs the code a million times (by default) to get the most precise value
# for the code execution time.

import timeit
import matplotlib.pyplot as plt

price = [2.50, 1.23, 4.02, 3.25, 5.00, 4.40]
sales_per_day = [34, 62, 49, 22, 13, 19]

print(
    "plt.scatter()",
    timeit.timeit(
        "plt.scatter(price, sales_per_day)",
        number=1000, # default is 1,000,000
        globals=globals(),
    ),
)
print(
    "plt.plot()",
    timeit.timeit(
        "plt.plot(price, sales_per_day, 'o')",
        number=1000,
        globals=globals(),
    ),
)
```

###### The plt.hist() function has many parameters to tune both the calculation and the plot display. We can see these parameters in action as follows:-



- **bins**, by default 10
- **range**, if not provided, range is (x.min(), x.max())
- **density**, If True, draw and return a probability density: each bin will display the bin's raw count divided by the total number of counts and the bin width (density = counts / (sum(counts) * np.diff(bins))), so that the area under the histogram integrates to 1 (np.sum(density * np.diff(bins)) == 1).
- **It's a well-known fact that the largest value a probability can take is 1**. when plotting density / pdf u will observe irrespective of actual y-values, they will be normalized such that area under curve integrates to 1.


- **weights**, An array of weights, of the same shape as x. Each value in x only contributes its associated weight towards the bin count (instead of 1).
- **cumulative**, If True, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values. The last bin gives the total number of datapoints.**If density is also True then the histogram is normalized such that the last bin equals 1** we know a largest value probability can take is 1.
- **histtype**{'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar'


**Distribution**

A distribution in statistics or probability is a description of the data. This description can be verbal, pictorial, in the form of an equation, or mathematically using specific parameters appropriate for different types of distributions. Statisticians have observed that frequently used data occur in familiar patterns and so have sort to understand and define them. **PDF**: Probability Density Function, returns the probability of a given continuous outcome. 

**CDF**: Cumulative Distribution Function, returns the probability of a value less than or equal to a given outcome.

##### Compare Histograms of several distributions


Using np.random.normal will generate values from normal distribution which 
will range from (mean - 3 std) to (mean + 3 std)<br>
Example if mean=1 and std=5 then values ranging b/w -14 to 16, 99.73% data will be covered.



```python
# np.random.normal to draw random samples from a normal (Gaussian) distribution.
# syntax - random.normal(loc=0.0, scale=1.0, size=None)¶where loc is mean/center of
# distribution, scale is std dev or width of distribution, size=no of samples


x1 = np.random.normal(0, 4, 1000)   # every run different value generated for random numbers
x2 = np.random.normal(-2, 2, 1000)
x3 = np.random.normal(1, 5, 1000)    
# 3s.d range for x3 should be (1-3*5 = -14) to (1+3*5 = 16) as per mean=1 & std=5
# 99.73% data should fall here.

# Check for any value outside of range
for i in x3:
    if (i > 16) | (i < -14):
        print('Outside 3 s.d range for x3',i)
        
plt.hist(x1, label='x1')
plt.hist(x2, label='x2')
plt.hist(x3, label='x3')
plt.legend();
    
```

Such a distribution is specified by its mean and covariance matrix. These parameters are analogous to the mean (average or “center”) and variance (standard deviation, or “width,” squared) of the one-dimensional normal distribution.


```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x8, y8 = np.random.multivariate_normal(mean, cov, 10000).T

df = pd.DataFrame({'x':x8, 'y':y8})  # creating df from dict
print(df.head())

plt.hist2d(x8, y8, bins = 30, cmap = 'Blues')

cb = plt.colorbar()

cb.set_label('counts in bin')
```

###  Stacked Bar Chart
We can draw stacked bar chart by using a special parameter called bottom from the plt.bar() function. It can be done as follows:-

```python
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)  # bottom parameter for stacked bar chart
plt.title('Stacked bar chart')

plt.show()
```

### Pie Chart
- Pie charts are circular representations, divided into sectors. 
- The sectors are also called wedges. 
- Matplotlib provides the pie() function to plot pie charts from an array X. 

```python
plt.figure(figsize=(5,5))

x10 = [35, 25, 20, 20]

labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']

plt.pie(x10, labels=labels);  # This is one dimensional chart only, just requires x.
plt.title('Simple Pie chart')

plt.show()
```

#### Exploded Pie Chart
-exploded Pie chart with the addition of keyword argument explode. 
- It is an array of the same length as that of X. Each of its values specify the radius fraction with which to offset the wedge from the center of the pie.

```python
plt.figure(figsize=(5,5))

x11 = [30, 25, 20, 15, 10]

labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical', 'Agriculture']

explode = [0.2, 0.1, 0.1, 0.05, 0]

plt.pie(x11, labels=labels, explode=explode, autopct='%1.1f%%');
# autopct='%1.2f%%' means that for each pie wedge,format string is '1.2f%', till 2 floating point.

plt.show()
```

### Boxplot
-Boxplot allows us to compare distributions of values by showing the median, quartiles, maximum and minimum of a set of values.
-  boxplot with the boxplot()

```python
# random.randn - Return a sample (or samples) from the “standard normal” distribution.

data3 = np.random.randn(100)

plt.boxplot(data3, showmeans=True)
plt.title('Boxplot')

plt.show();
```

The boxplot() function takes a set of values and computes the mean, median and other statistical quantities. The following points describe the preceeding boxplot:

• The red bar is the median of the distribution.

• The blue box includes 50 percent of the data from the lower quartile to the upper quartile. Thus, the box is centered on the median of the data.

• The lower whisker extends to the lowest value within 1.5 IQR from the lower quartile.

• The upper whisker extends to the highest value within 1.5 IQR from the upper quartile.

• Values further from the whiskers are shown with a cross marker.

#### Customized Boxplot
- We can draw several boxplots with just one single call to plt.boxplot() as follows:

```python
data4 = np.random.randn(100, 5)  # creates an array of shape (100,5) & fills with random
                                 # values as per standard normal distribution.
plt.boxplot(data4)

plt.show();
```

### Area Chart - Showing trend with magnitude - like line+bar chart
- An Area Chart is very similar to a Line Chart.
- The area chart is like the plot/line chart except that the area below the plotted line is filled in with color to indicate volume. Hence Area Chart shows magnitute with trend.
- The area between the x-axis and the line is filled in with color or shading.
- An area chart is a good way to demonstrate trends over time to the viewer, for example showing the total population over time. This chart is based on the line chart. 

- Built using fill_between function
- An area chart combines the line chart and bar chart to show how one or more groups’ numeric values change over the progression of a second variable, typically that of time.
- A good example is tracking changes in sales of different products a company sells over time, with the top line showing total sales.

```python
# Create some data
x12 = range(1, 6)
y12 = [1, 4, 6, 8, 4]

# Area plot
plt.fill_between(x12, y12)
plt.title('Area plot')
plt.show()
```
