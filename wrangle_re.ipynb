# Regular Expressions

- A regular expression also called as regex is a sequence of characters that define a search pattern.
- To simplify searching operation in string.
- Use regular expressions to extract patterns from data.
- Use regular expressions to clean bad data.
- Most of languages support regular expressions & patterns used r almost same. Patterns r language agnostic.
- Very complex searching patterns can be easily done using simple regex expression.

**Note**

Whitespace is also a character according to any tool.
