## EDA - Exploratory Data Analysis - main objective is data discovery!
 
### You get to know interesting and non-intuitive information about your data. This helps to tell compelling stories from the analysis based on which many Business Intelligence dashboards r developed which are used for taking data backed business decisions.



### Definition

1. EDA is an approach to analysing datasets to summarize their main characteristics, often with visual methods.
2. EDA techniques r either graphical or quantitative(non-graphical).

### Uses of EDA

1. Understand data quality/sanity.
2. Understand relationships b/w variables.
3. Used for feature selection in ML.
4. For building useful dashboards.
5. For storytelling.

- The first 2 points r considered as data preprocessing steps while building ML/DL models.


### How to perform EDA?

Ask questions- multiple questions, different questions. It is a powerful skill to do EDA.

### Where EDA could be helpful?

1. To predict what is probability of an employee leaving an organisation? -> ML involved since predictive analysis
2. To identify/understand the parameters which influence the increase/decrease in revenue? -> no ML involved descriptive analysis

LHS for use case1 and RHS for use case2!

### Points to remember before starting EDA

1. Identify the parameter which is of focus. The parameter which is of focus becomes our **Key Point Indicator (KPI).** 
2. E.g. in LHS problem above, attrition(employee left or not) is KPI and in RHS problem Sales/Revenue is KPI.
3. EDA techniques to understand what influences our KPI- Always start with **Descriptive statistics** to understand meta information about our data. e.g. Data size, format, data types & summary of each column.This will help to choose right tool for analysis.



### Some EDA techniques to analyse data


1. **`Pareto Analysis`** to identify vital few which influences your KPI from trivial many.
2. **`Distribution Analysis`** if KPI is numerical like sales/profit etc can do distribution analysis to see how KPI is distributed.
3. **`Outlier Analysis`** identify any outliers which may drive your KPI
4. **`Trending Analysis`** if u have past data can do trending analysis to understand how KPI is changing across time e.g. seasonality, trends etc
5. **`Geographical Analysis`** to understand how KPI's are distributed across various locations.
