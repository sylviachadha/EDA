# Parameters

When u want to dynamically adjust visualization, e.g. change size of bins in age or income distribution, we can use Parameters.

Specifically in data exploration & data discovery u need to be agile, to quickly adjust things to get different visualizations & help find anomalies/insights etc


Parameter is created and then do show parameter.

You can see the parameter now and this needs to be linked to the field balanced bins whose values u want to adjust dynamically.

Click on field to which parameter needs to be linked, in this case Balance bin.

And then choose Balance Groups parameter which is already created.

Now when u select from Balance Group parameter, the size of chart adjusts accordingly. 5000, 15000 and 25000 bin size and charts change accordingly.
