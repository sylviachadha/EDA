## Data Blending

- Data blending is a method for combining data from multiple sources.It is used when there is related data in multiple data sources, which you want to analyze together in a single view.
- The two sources involved in data blending are referred to as primary and secondary data sources.
- A left join is created between the primary data source and the secondary data source with all the data rows from primary and matching data rows from secondary data source. It is a kind of smart join.
- Data blending is required when granularity (aggregation level) of data in tables is different.
