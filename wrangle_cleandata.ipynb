### Data Quality Issues specific to data and domain

These data quality issues r specific to this dataset & this domain. Not all datasets will have same issues. Each dataset will have its own pattern of issues that cannot be identified automatically, hence manual cleaning required.




1. Country column - Japan **misspelt**. To check this use unique counts & see if any categories look incorrectly spelt.
2. Sales column cz of currency symbol 'dollar' and ',' it is interpreted as string. So need to remove **special characters & comma** & typecast this column into numeric. We will not be able to perform any numerical operation on top of string column.
3. **Multiple currency symbols** e.g. 'dollar' and INR in same column.
4. There r **missing values** in Sales and Volume Column given by NA, - or completely blank. Need to do missing value treatment.
5. In date column, **inconsistent Date format**
6. Discount **range** column can be split with regular expressions in lower discount & upper discount so we can do some numerical analysis on each of 2 col's. In 1 row only flat 5% discount. Can put in both upper & lower, depends on business logic.
7. In num of Outlets col a string s with 100 in 1 record makes whole col as string. Need to detect such and convert to numerical.


### Common occuring Data Quality Issues

### Best way to avoid data quality issues is to store data in database tables where data types r predefined for each column, before populating values.

While inserting records to a table which has predefined schema, if there r any values which do not abide corresponding columns data type then tool will throw an error.
