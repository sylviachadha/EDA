# Tableau

**What is Tableau?**


Tableau tool to work with data. See, understand and visualize data. Tableau is a **drag and drop software.** Visuals can be created faster in tableau compared to in other pgms/software. Tableau is a new & innovative approach to **business intelligence.**

Connect to various data scources and perform Aggregations on data. Make **Dashboard** with maps and scatterplots, interactive filters. Joining and Blending concepts in tableau. Customer segmentation dashboard. Creating a Storyline. Clustering analytics in tableau.

**What is tableau used for?**

Tableau is a business intelligence and data visualization tool to get insights from data, find hidden trends, and make business decisions
