## Tableau - Introduction to data Visualization

**1. Data Visualization**

Graphical representation of information and data. Two basic types of data visualization r static and interactive.

**2. Need for Data Visualization**

- Data is growing v rapidly. Just storing data will not help, to consume such large amount of data we need modern visualization tools to explore data & understand its characteristic. 
- Drill down deeper with high granularity.
- Through visuals easy to summarize characteristic of data & identify patterns in data.
- Detecting outliers helps organisations to identify fraud. (Boxplots)
- Instead of understanding data by just numbers and text, it is easy to consume data through visuals & hence communicate results better & faster to all stakeholders. Strategic decisions taken using data visualizations.

**3. Classification of visuals**

1. **Infographics** - Visually appealing, Story-driven, handcrafted but less reproducibility.

2. **Data Visuals** - Visually informative, data-driven, easy to refresh with new data. Some egs of data visuals r bar charts, line charts & scatter plots.

3. **Exploratory** - In many visuals, filters are added for drill down, so can play around & come up with ur story. 
4. **Explanatory** - These r static or animated. Message is conveyed to convince the user. Msg's could be biased.
5. **Operational Dashboards** - To monitor current state of affairs. Data is periodically updated & these visuals will be typically in the form of dashboards.
6. **Analytical Visuals** - These r generally used to summarize results from ML or any analytical models. Once summarized the data is very rarely updated.
