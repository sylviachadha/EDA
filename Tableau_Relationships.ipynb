# Relationships

- Relationships r more flexible than joins.
- Tables r treated as seperate when they r connected into relationships.
- If in doubt to use b/w join, blend, relationship, use a relationship.

## Dataset

https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce

Below is a kind of a data model (how different tables are connected together).

- If we have to do this data model with joins and blends it is difficult while with relationships it is easier.
- Joins, u can't always join if level of aggregations of 2 datasets r different then u have to use a blend.
- A blend can only be created per sheet.
- All this can be done easily with concept of relationships in tableau.

When u do join 2 tables r connected together(merged) but v want to have flexibility as seperate tables/db's & still able to call right information when we need it. Create relationships in logical layer only, not go to physical layer like in joins. You connect to seperate tables just there is a relationship b/w tables.

In case tableau can't automatically detect relationship then can edit it yourself.

# Data Model

A data model consists of a logical layer and a physical layer.

Relationships in logical layer but each table will have its own physical layer.

You can also use joins with relationships

Instead of relationship of olist_customer with geolocation, you can also create a join instead.

Now its a join instead of relationship-

If there might be different levels of aggregation or u just need seperate tables then use relationships while in some cases joins may be obvious where no complexity.

In most cases relationship will work well.
