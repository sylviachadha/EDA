## Distribution Analysis techniques

## Distribution Visualization techniques for Numerical Columns - Most of ML models expect normally distributed data hence need to do distribution analysis for presence of skew if any.

# 1. Boxplots

**Interpreting Distribution of Numerical KPI using Boxplot**
1. Minimum
2. Maximum
3. Mean
4. Median
5. Quartiles
- IQR = Q3-Q1
- Lower Whisker = Q1-1.5*IQR
- Upper Whisker = Q1+1.5*IQR
- Any value below lower whisker & above upper whisker is statistically considered as an outlier.

### Example - Product Rating Across Cities
- To check if Product Rating KPI influenced by categorical variable city. Plot different boxplots for each city.
- IQR range is small for Chennai and position of IQR on lower side of Product rating hence Consistent Poor rating in by customers in Chennai.
- IQR range is small for Bangalore however position of IQR on upper side of Product rating hence Consistent Good rating by customers in Bangalore.
- IQR range is big in Mumbai which means ratings by customer is varying. Mix of opinions in this city - product good and product average so further check which customer segment dislikes product & change business strategy accordingly.
- In case of Mumbai median is not in center of box which means data is not normally distributed & skewed.

### Normally Distributed Data as seen from Boxplot

- If median in middle of box and Q3-Q2 appx equal to Q2-Q1 then we say Normally distributed and no skew in data.
- If data is skewed mean is an unreliable metric.

### Left skew or Negative Skew

If median towards right of box and Q3-Q2 is less than Q2-Q1 then we say left/-ve skewed data.


### Right skew or Positive Skew

If median towards left of box and Q3-Q2 is greater than Q2-Q1 then we say right/+ve skewed data.

## Why detecting skew is important w.r.t ML ?

- Detecting skew is important cz many algorithms expect data to be normally distributed. If this is not the case the results from the model might not be reliable.
- Right skew w.r.t tail
- Left skew w.r.t tail

### How to overcome Skew ?

- Square root transformation compresses higher values, so lower values become more spread out. Log transformation compresses high values more aggresively than the square root transformation.

### This is one of feature engineering technique before building any ML model!

# 2. Histograms

- Apart from boxplot we can also use histogram to understand distribution of numerical variables. For histogram we need binned data.
- The difference b/w barplot & histogram is that in a barplot u have discrete/categorical variable on x-axis while histogram has continuous bins and bars have no spaces in b/w.

- From histogram we can see most customers age lie b/w 40 to 50.
- Histograms can be used to detect skewness in a numerical variable.
- Histograms can be normally distributed, left skewed or right skewed & skew is always w.r.t tail.


- Apart from detecting skewness we can also detect outliers.
