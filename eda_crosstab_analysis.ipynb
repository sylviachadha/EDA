## Crosstab Analysis - eda technique

Crosstab Analysis to understand our KPI metrics.

- Analysis on long format data is difficult so we pivot the data and change to wide format.
- LHS is long format
- RHS is wide format

### Definition

- This way of analysing by converting long format data into wide format data is known as CrossTab Analysis.
- The heatmap interpretation to understand Sales value- dark color high sales and v.light color meaning least sales.

**Analysis w.r.t wide format**

To understand how cities & products r influencing Sales values.

1. How single product in selling in various cities e.g. Product1 is performing good in all cities except Delhi.
2. How different products r selling in each city e.g. in city of bangalore sales value of most products good except Product4.

### When to use Crosstab Analysis

- When u want to analyse Categorical vs Categorical columns using frequency as metric.
- When u want to analyse Categorical vs Categorical columns using numerical KPI.

### Use of Crosstab Analysis

- To understand dependency b/w 2 categorical columns.
- To understand how a KPI is influenced by two categorical variables.
- To detect outliers.

### Ways to visualize Crosstab Analysis -
1. Heatmap
2. Stacked Bar Chart
