# Custering & Trend lines in Tableau

Tableau applies K-means Clustering algorithm.

In the Analytics tab we have the clustering option.

- Clusters based on 2 variables & clusters based on 3 variables. When a 3rd variable is added due to its effect the clusters change. As we keep on adding variables tableau would group them in multidimensional space - 4D, 5D ,10D space.
- Although below we can observe only w.r.t 2 dimensions but tableau is calculating proximity of observations based on values of all variables in n-dimensional space.
- Forming groups of proximity of these observations.

Above Clustering shows that apart from Marketing spend, population also effects the revenue hence clusters change after adding 3rd variable of population.

### Analysing Trends in different clusters

In the Analytics tab u have the trends line.



Adding Trend lines on Clusters

For below red line equation states that for every dollar spent on marketing u get 94 cents of revenue. The slope coefficient is less than 1, basically u r losing money.

For Orange line, for every dollar invested on marketing spend u r getting an increase of 7 dollars on revenue.

For blue line, for every dollar invested on marketing spend u r getting an increase of 3 dollars on revenue.
