# Time Series

Tableau highlights dimensions in blue and measures in green.

When we do time series we want time 'as a timeline' instead of 'categories'. Below in LHS chart since year appears in blue color means tableau is taking it as dimension/category/qualitative data while in RHS year appears as quantitative data/measure.

You need to decide based on your requirement whether u want time as a category or as a measure. 

## Viewing data at different granularities.

In markers can choose granularity to see dots for each data point.

If u want to plot at different levels of granulaity u would want to have time as a measure.

By year, Quarter and Month - level of granularity more & hence more data points.

## Aggregation (Measures), Granularity(Dimensions) and Level of detail

**Measures get aggregated and dimensions specify level of granularity.**

By default Analysis has aggregate measures, u can also uncheck it.

LHS with aggregation (default) while RHS without aggregation (will plot all raw data points/rows) - yearly & monthly.

Introducing dimension will change level of granularity and therefore will affect the aggregation.

## Adding granularities

e.g. adding gender in level of granularity, below fig. based on granularity of month and gender.

We can add more dimensions to increase granulaity e.g. also adding age to shapes tab.

## Different aggregations

There r different aggregations available like Sum, Average, Median, Count etc

## Detail

Detail tab as shown below. It allows you to add measures and dimensions to your dashboard without affecting visual representation but affecting level of granularity.

Sometimes u want to increase your level of granularity, u want to make ur dashboard more detailed but u do not want to drag anything on your color, shape. u do not want to affect the visual by adding dimensions to color, shape but still want to add level of granularity, in that case can use DETAIL tab. Detail refers to level of granularity of your chart.

When add Age to level of detail, the shape has not changed but level of granularity has still changed.
