## Pareto Analysis - EDA Technique
### Identify & seperate vital few from trivial many

Based on -
1. dataset, 
2. types of columns
3. data type of your KPI 

u have to decide which EDA technique is appropriate.

### PAreto Definition

1. It is a technique used in business decision making based on 80/20 rule.
2. 80% of reward comes from 20% of vital input.
3. e.g. 80% revenue comes from 20% products so those 20% products r important as they make up 80% of revenue.
4. Hence those 20% r vital influencers for our KPI.
5. These KPI's could be Sales, Volume, Efficiency, Profit, Transactions, Downloads etc.

### When to use Pareto Analysis ?

1. In a categorical column to identify important categories that drive the KPI.
2. Generally we will have 1 categorical and 1 numerical KPI.
3. In case there is only 1 categorical column in that case use freq of category as a metric.

### Pareto Chart

- When we combine above 2 charts in a single chart it is called as a Pareto Chart.
- The bars indicate absolute sales values which r shown on left side y-axis.
- The line chart indicates cumulative % of sales of each product & for this use right side y-axis.
- A pareto Chart contains both bars and a line chart where individual values r represented in descending order by bars & the cumulative total is represented by the line.


### How Pareto Chart useful w.r.t ML?
1. When building ML models certain categorical columns having many categories might create an issue.
2. In such cases we can use Pareto Analsysis and identify those 20% categories which r important & retain them.
3. Club remaining 80% categories(trivial ones) into a single category called as others.
