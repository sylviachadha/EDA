## Descriptive Statistis for EDA

The first few questions that need to be asked even before looking at your data that would help u decide which tools u need to use, to analyse your data.


### **Size of data**
- upto few gb's just load data in python & use pandas to do eda.
- if greater than 5 gb's, ram might not be sufficient to load all data in memory hence python & pandas alone can't be usd. Load data in database & use combination of sql & pandas operations to perform eda.
- if data v big, big data solutions to do some analysis on entire dataset.

### **Missing data** 
- If percentage of missing values very high, results might not be reliable in such cases.

### **Data types**
- Identifying what type of data u have is crucial to identifying right kind of techniques.
- Typical data types will be - Numeric, Categorical/Levels/Groups/Dimensions, Texts - Text analysis, Dates -Trend analysis, Geographical - geographical analysis using maps.
- Other Columns like phone no's, email ids, URL might be useful for datawrangling but might not always be useful for data analysis.
- After identify data type of each column, summarize individually (each column) to understand data at a high level.
