## Subsetting- selection, slicing, indexing, filtering
We use square brackets [] to select a subset of an Python object.

**Selecting**<br>
Selecting data using Column headings.


**Slicing** possible on:
- Range based Subsets
- Subsets of Rows
- Subsets of Rows and Columns<br>
loc is primarily label based indexing<br>
iloc is primarily integer based indexing<br>

Slicing generally extracts sequence of elements.<br>

**Indexing**<br>
Indexing is used to obtain individual elements


**Filtering**<br>
Extraction based on conditions

### 1. Subsetting - Selection

```python
# Column selection
df['product_cat']
```

```python
# Column selection
df.product_cat
```

```python
# Multiple Column selection
df[['product_cat','sales_Qty','revenue']]
```

### Difference b/w [] and [[]]



- df['product_cat'] returns a series (1 Dimensional)


- df[['product_cat','sales_Qty','revenue']] returns a dataframe (2 Dimensional)

### 2. Subsetting -Slicing (  Slicing Operator ':' )

The main difference between pandas loc[] vs iloc[] is loc gets DataFrame rows & columns by labels/names and iloc[] gets by integer Index/position.

```python
# Selecting a range of rows from the DataFrame
# selecting range of rows from 2 to 5
display(df.loc[2 : 5])
```

```python
# Slicing all rows and only column 'sale_date'

df.loc[:, ['sale_date']]
```

```python
# Selecting 0th, 2th and 4th index rows

display(df.iloc[[0, 2, 4]])
```

```python
# selecting rows from 1 to 4 and columns from 2 to 4
display(df.iloc[1 : 5, 2 : 5])   # since upper bound not included in python
```

```python
df.iloc[:,[0,2]]   # display all rows and columns 0,1
```

### 4. Subsetting - Filtering

Selection based on conditions (can use .query) to query df based on condition

```python
df
```

**Question** - Retrieve all Z category products

**Solution** - This can be achieved through 3 methods as below-

```python
# Method 1
df[df['product_cat']=='Z']
```

```python
# Method 2
df[df.product_cat=='Z']
```

```python
# Method 3
df.query("product_cat == 'Z'")
```

**Question** Retrieve all Y and Z category products

**Solution**

```python
# Method 1
df[(df['product_cat']=='Y') | (df['product_cat']=='Z')]
```

```python
# Method 2
df[(df.product_cat=='Y')|(df.product_cat=='Z')]
```

```python
# Method 3
df.query("(product_cat =='Y') or (product_cat =='Z')")
```

**Question** - Retrieve sale date for all Y and Z category products

**Solution**

```python
# Method 1 put condition in () and then column u want information on in [[]]

df.query("product_cat == 'Y' | product_cat == 'Z'")[['product_cat','sale_date']]
```

```python
# Method 2 

#df[(df['product_cat'] == 'Y') | (df['product_cat'] == 'Z')]

df[(df['product_cat'] == 'Y') | (df['product_cat'] == 'Z')][['product_cat','sale_date']]
```

**Question** - Retrieve sale_date, sales_qty, revenue for products X and Y

**Solution**

```python
df.query("product_cat == 'X' | product_cat == 'Y'")[['sale_date', 'sales_Qty','revenue']]
```

```python
df[(df.product_cat=='X')|(df.product_cat=='Y')][['product_cat','sale_date','sales_Qty','revenue']]
```

**Question** - selecting product_cat=X  and sales Qty > 50

**Solution**

```python
df[(df.product_cat == 'X') & (df.sales_Qty > 50)]
```

**Question** - sales Qty > 50

**Solution**

```python
# Method 1

print(df['sales_Qty'] > 50)

print('\n',df[df['sales_Qty'] > 50])
```

```python
# Method 2

df.query('sales_Qty > 50')
```

**Question** - product_cat where sales Qty > 50

**Solution**

```python
# Method 1

df[df['sales_Qty'] > 50][['sales_Qty','product_cat']]
```

```python
# Method 2

print(df.query("sales_Qty > 50")[['product_cat','sales_Qty' ]])  # If sales Qty > 50, then print product_cat
```

### Aggregating Data

combining individual-level data

Aggregating data is when u r trying to group data based on some function.<br>
It is different than retrieving data (which we call as filtering). U r just filtering data based on some conditions and not trying to apply some function and summarize it.



```python
df
```

**Question** -

Get the sales quantity sum for each catagory


```python
# Error result

DF1=df.groupby(['product_cat']).sum()
DF1
# What is the error here? - It sums all the numerical columns leaving categorical, datetime cols
```

```python
# Error result

DF2=df.groupby(['product_cat']).agg('sum')
DF2
# What is the error here? - It sums all the numerical columns leaving categorical, datetime cols
```

### Aggregate function - Sum

#### Method 1 

```python
# Method 1 - this returns a dataframe but groupby column is made as index

DF=df.groupby(['product_cat']).agg({'sales_Qty': 'sum'})
DF
```

```python
# Above operation made product_cat as the index instead of column name as shows only
# sales_Qty as column name.

print(type(DF))
print(DF.columns)
print(DF.index)

# Product catagory filtering & subsetting is not possible for DF, need reindexing now
```

```python
df.groupby(['product_cat']).agg({'sales_Qty': 'sum'}).reset_index()
```

#### Reset index and rename column for aggregated result for understanding of what column conveys
It conveys aggregated sales value now instead of individual sales

```python
df.groupby(['product_cat']).agg({'sales_Qty': 'sum'}).reset_index().rename(columns={'sales_Qty':'total_sales'})
```

#### Method 2

```python
# Method 2 - this returns only a series

DF1 = df.groupby('product_cat')['sales_Qty'].sum()   # if groupby needed on more than 1 column
print(DF1)                                           # then pass as list df.groupby(['A','B'])

# What column do we want to group by - groupby(['Item_Type'])
# What column we want information returned on - ['Item_MRP']
# What type of information we want returned - e.g. sum / mean etc
```

```python
print(type(DF1))
#print(DF1.columns) - Not Applicable for series
print(DF1.index)
```

```python
# So to get product_cat as column and not as index i.e. series just reset index-

DF1 = (df.groupby('product_cat')['sales_Qty'].sum()).reset_index()
print(DF1)
print(DF1.columns)
```

### Aggregate function count()

```python
# Method 1
df.groupby(['product_cat']).agg({'sales_Qty': 'count'}).reset_index().rename(columns={'sales_Qty':'count_sales'})
```

```python
# Method 2 - Instaed of using sales_Qty can just use product_cat as we just need count which
# is not related to sales_Qty.
# But here then reset_index() will not work as column name product_cat same.

DF1 = df.groupby('product_cat')['product_cat'].count()
print(type(DF1))
DF1
```

```python
# So to use reset_index() u will have to give a different column name

DF2 = (df.groupby('product_cat')['sales_Qty'].count()).reset_index()
print(type(DF2))
DF2
```

```python
# Method 3
# For count of product_cat can simply use value_counts instead of groupby

ans = df['product_cat'].value_counts()
print(type(ans))
ans
```
