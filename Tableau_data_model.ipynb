# Data Model

3 tabs in excel appear as 3 tables.

## **Physical Layer** (Joins)

Physical layer of tableau.At physical layer, tables are connected via joins. As shown below, it says double click logical table to see its physical table.

After u double click, u r in physical layer.

In physical layer if u connect tables u will have a join.

There can be different types of joins, here we have inner join.

By joining tables u r not changing data, u r just changing structure of data, now fields from 2 tables appear together in joined table.

## **Logical Layer** (Relationships)

There is a logical layer for tables. At the logical layer the tables r connected with relationships.
