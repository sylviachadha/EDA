# Joins

- Joins r operations performed on tables in relational databases (mysql, postgres, mariadb etc) where data is stored in a structured format.
- Inner joins and left joins r used more frequently.

**1. Inner Join** - Rows common in both left and right tables.

**2. Left Join** - Primary table on left. Rows common in both left and right tables + remaining rows of left table. Since left table is primary table, rows can't be discarded from it, rows can only be discarded from non-primary table.

**3. Right Join** - Primary table on right. Rows common in both left and right tables + remaining rows of right table. Since right table is primary table, rows can't be discarded from it, rows can only be discarded from non-primary table.

**4. Full Join** - Matched rows in both tables + All remaining rows in both left and right tables. This join will not discard any of the rows in any of the 2 tables.

# Duplicates in Joins

When one of the table has duplicate values e.g. 1 table has order number and the other table has itemized order details & hence order number is duplicated in this table.

# Joining on multiple fields

If values in a particular column r not unique then that column cannot be identified as key or unique identifier & joining them based on that column can lead to incorrect results. In that case we may need to make 2 columns combination as a key.
