### rc.Params

- *<font color=blue>By modifying plt.rcParams, you can customize the look and feel of your plots globally, so that all subsequent plots will adhere to these settings without needing to specify them each time you create a plot.*

- In rcParams, "rc" stands for "run commands." The term comes from the concept of "run commands" in configuration files, which are often referred to as "rc files" in Unix and Linux systems. These files typically contain settings and configurations that are read and applied when a program starts. Similarly, in Matplotlib, rcParams contains the settings and configurations that determine the default behavior and appearance of plots.

- plt.rcParams is a dictionary-like object in the Matplotlib library in Python that contains configuration settings for customizing the appearance of plots. These settings, often referred to as "rc settings" or "rc parameters," control the default styles and behaviors of Matplotlib plots.

You can use plt.rcParams to change a wide range of settings, such as:

- Figure size and DPI - "dots per inch" it is a measure of the resolution of a figure. It determines how many pixels per inch are used to render the figure. 
- Line styles and widths
- Font sizes and styles
- Colors and color maps
- Axes properties
- Ticks and labels
- Legends and grids


```python
plt.rcParams['figure.figsize'] = [8,4]
plt.rcParams['figure.dpi'] = 140

# 140 dots per inch (DPI). This means that for every inch of the figure, there will be 140 pixels. 
# This affects the clarity and size of the figure when displayed or saved.
```

# 2. Heatmap

Heatmaps can be useful for visualizing time series data, especially when you want to show how a particular value changes over time in relation to two dimensions, such as time of day and day of the week, or year and month.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulate 10 years of monthly data
dates = pd.date_range(start='1/1/2010', end='12/31/2019', freq='M')
values = np.random.rand(len(dates))  # Random values for demonstration

# Create DataFrame
df = pd.DataFrame({'date': dates, 'value': values})
df.set_index('date', inplace=True)

# Add 'Year' and 'Month' columns
df['Year'] = df.index.year
df['Month'] = df.index.month

# Pivot the DataFrame to create a matrix for the heatmap
heatmap_data = df.pivot_table(values='value', index='Year', columns='Month', aggfunc='mean')

# Create the heatmap using Matplotlib
fig, ax = plt.subplots(figsize=(12, 8))
cax = ax.matshow(heatmap_data, cmap='YlGnBu')

# Add colorbar
fig.colorbar(cax)

# Set axis labels
ax.set_xticks(np.arange(heatmap_data.columns.size))
ax.set_xticklabels(heatmap_data.columns, rotation=90)
ax.set_yticks(np.arange(heatmap_data.index.size))
ax.set_yticklabels(heatmap_data.index)

# Set labels
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Heatmap of Values by Year and Month (2010-2019)')

# Show the plot
plt.show()
```

# 3. Boxplot

Purpose: Shows the distribution and outliers over time or grouped by periods.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulate 10 years of monthly data
dates = pd.date_range(start='1/1/2010', end='12/31/2019', freq='M')
values = np.random.rand(len(dates))  # Random values for demonstration

# Create DataFrame
df = pd.DataFrame({'date': dates, 'value': values})
df.set_index('date', inplace=True)

# Add 'Month' column
df['Month'] = df.index.month

# Prepare data for box plot
data = [df[df['Month'] == i]['value'].values for i in range(1, 13)]

# Create the box plot using Matplotlib
plt.figure(figsize=(12, 6))
plt.boxplot(data, labels=range(1, 13))
plt.title('Box Plot of Values by Month')
plt.xlabel('Month')
plt.ylabel('Value')
plt.show()
```
