## Data Fundamentals


**Data**<br>
- Data can be defined as raw facts generally represented by values which can be recorded but have little meaning. Data cannot be used in unprocessed state for decision making.

- Data is factual information about an object. The object can be tangible or intangible. e.g. Bank customer is a tangible object while Bank Account is an intangible object. 


**Information**<br>
When raw facts(data) are processed based on business logic using a software program to reveal their intrinsic meaning.Then can be used for any meaningful purpose / decision making.

**How Computer system interprets data at hardware level?**

- Computers work on electricity. Computers are digital machines & computer hardware has been designed to handle & process data only in digital form.Digital signals have non-continuous electrical signals while analog signals have continuous electrical signals.


- Every computer has a microprocessor (CPU - Central Processing Unit) which performs all arithmetic and logical operations. e.g. Ryzen is AMD processor, M1 is ARM processor, then we have Intel(x86) processors. (ARM is RISC) while (AMD and Intel(x86) is CISC). CISC and RISC (Complex and Reduced Instruction Set Computer) are dominant processor architecture paradigms.


**CPU**

CPU is made up of millions of tiny circuits called transistors. These transistors inside processor chip function as micro switch that can be programmed to be switched 'on' and 'off' These 2 states r represented using binary number system which makes use of only 2 numbers 0 and 1.

**ASCII Value**<br>

Cmmputers convert text and other data into binary with an assigned ASCII value. Once ASCII value is known, that can be converted into binary.
