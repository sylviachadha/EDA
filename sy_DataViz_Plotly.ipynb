# Line Plot vs Area Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = {
    'Year': np.arange(2000, 2010),
    'Value1': np.random.randint(1, 100, 10),
    'Value2': np.random.randint(1, 100, 10)
}
df = pd.DataFrame(data)

print(df.head())

# Line plot
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Value1'], marker='o', label='Value1')
plt.plot(df['Year'], df['Value2'], marker='o', label='Value2')
plt.title('Line Plot Example')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Area plot
df.set_index('Year', inplace=True)
df.plot(kind='area', alpha=0.5, figsize=(10, 5))
plt.title('Area Plot Example')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

# 7. Spread Plot

*<font color=blue>Here it is the mathematical difference. This is different from statistics. In statistics, "spread" generally refers to measures of dispersion or variability like range, IQR, variance, and standard deviation.*


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create dummy data
np.random.seed(42)
df = pd.DataFrame({
    'Series1': np.random.randn(100).cumsum(),
    'Series2': np.random.randn(100).cumsum()
})

# Display the first few rows of the DataFrame
print(df.head())

# Calculate the spread (difference between Series1 and Series2)
df['Spread'] = df['Series1'] - df['Series2']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Spread'], label='Spread (Series1 - Series2)', color='blue')
plt.fill_between(df.index, df['Spread'], where=(df['Spread'] >= 0), interpolate=True, color='green', alpha=0.3, label='Positive Spread')
plt.fill_between(df.index, df['Spread'], where=(df['Spread'] < 0), interpolate=True, color='red', alpha=0.3, label='Negative Spread')
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Spread Plot Example')
plt.xlabel('Index')
plt.ylabel('Spread')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Create dummy sales data for two products
dates = pd.date_range(start='2023-01-01', periods=100)
sales_a = np.random.randint(50, 200, size=100).cumsum()  # Simulate cumulative sales for Product A
sales_b = np.random.randint(50, 200, size=100).cumsum()  # Simulate cumulative sales for Product B

# Create a DataFrame
df = pd.DataFrame({
    'Date': dates,
    'ProductA Sales': sales_a,
    'ProductB Sales': sales_b
}).set_index('Date')

# Calculate the spread (difference between ProductA Sales and ProductB Sales)
df['Spread'] = df['ProductA Sales'] - df['ProductB Sales']

# Display the first few rows of the DataFrame
print(df.head())

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Spread'], label='Spread (ProductA Sales - ProductB Sales)', color='blue')
plt.fill_between(df.index, df['Spread'], where=(df['Spread'] >= 0), interpolate=True, color='green', alpha=0.3, label='Positive Spread')
plt.fill_between(df.index, df['Spread'], where=(df['Spread'] < 0), interpolate=True, color='red', alpha=0.3, label='Negative Spread')
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Spread Plot Example: ProductA Sales vs ProductB Sales')
plt.xlabel('Date')
plt.ylabel('Spread')
plt.legend()
plt.grid(True)
plt.show()
```

# 9. Bubble Plot

*Bubble plots are particularly useful in scenarios where you need to visualize and compare relationships between three or more variables simultaneously, with an emphasis on the magnitude of data points. Example- Customer Segmentation: Visualizing customer segments based on demographics (e.g., age, income) and purchase behavior (e.g., total spend), with bubble size representing the number of customers. This helps in identifying which segments are most lucrative, which is harder to convey with just bar or scatter plots. Example- Epidemiology: Tracking the spread of diseases with location coordinates (x and y), number of cases (bubble size), and severity or mortality rate (color). This comprehensive view helps in understanding the impact and spread, which traditional heatmaps or scatter plots might not convey as effectively*

### Sample Problem

**Problem:**

You are a public health analyst tasked with monitoring the spread of a contagious disease across different cities. You need to visualize the number of cases, the severity of the disease (e.g., hospitalization rate), and the population of each city to identify hotspots and allocate resources effectively.

**Solution:**

A bubble plot can effectively show the relationship between these three variables, where:

X-axis: Number of Cases
Y-axis: Hospitalization Rate (or Severity)
Bubble Size: Population of the City
Bubble Color: Geographic Region or Alert Level

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Sample data generation
np.random.seed(42)
data = pd.DataFrame({
    'City': [f'City {i}' for i in range(1, 21)],
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 20),
    'Cases': np.random.randint(100, 5000, 20),
    'Hospitalization Rate': np.random.uniform(0.01, 0.2, 20),
    'Population': np.random.randint(50000, 1000000, 20)
})

# Display the DataFrame
print(data)

# Create the bubble plot
fig = px.scatter(data, x='Cases', y='Hospitalization Rate', size='Population', color='Region', 
                 hover_name='City', title='Epidemiology: Tracking Disease Spread',
                 labels={'Cases': 'Number of Cases', 'Hospitalization Rate': 'Hospitalization Rate'})

fig.show()
```

# 10. Heatmap

- cf.datagen.heatmap: This function generates random data suitable for creating a heatmap.
- Parameters (20, 20): This specifies the dimensions of the heatmap, i.e., a 20x20 grid.


```python
data1 = cf.datagen.heatmap(20,20)
data1
```

```python
data1.iplot(kind='heatmap', colorscale='spectral');
```

```python
# Create the heatmap
fig = data1.iplot(kind='heatmap', colorscale='spectral', asFigure=True)

# Customize y-axis to show every point
fig.update_layout(
    yaxis=dict(
        tickmode='linear',
        dtick=1
    ),
    title='Heatmap Example with Custom Y-Axis Ticks'
)

# Show the plot
fig.show()
```
