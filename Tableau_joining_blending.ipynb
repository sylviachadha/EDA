# Data Blending 



**Use blending when-**
1. If granularity of tables/datasets different, use blending.
2. If data sources different, use blending.

Use blending when level of granularity of two datasets is different as in that case we cannot perform a join.

We prefer blend than join in above case cz if the Sales table is also inner joined with above table we would lose/duplicate a lot of information cz the level of granularity is different.

If the data sources are different (e.g. you have 2 different excel files, or a file and a database) then you cannot do join in tableau in that case also there has to be blending of data.

### **Data Blending is a smart left join**

Blending is done on per sheet basis.

Blending is a smart left join as in sheet 1 table1 is primary table and blending happens as per that (6 rows) and in sheet 2 table2 is a primary table & hence blending happens as per that (4 rows) as shown below.

In data blending the data is first aggregated and only then joined in final result.

# Data Blending vs Joining 

When u join in tableau it happens at row level but when u blend it happens at granular level. Hence aggregation happens before the blend. Blends r unique to each worksheet. You can have 1 type of blend on 1 worksheet, other type of blend on another worksheet.

## Summary

- Use joins when combining data at row level
- Use Blends when datasources have different levels of granularity or use when datasources come from different systems.
